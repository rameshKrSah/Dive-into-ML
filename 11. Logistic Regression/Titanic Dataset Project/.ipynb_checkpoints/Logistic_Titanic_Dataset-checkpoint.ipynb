```python
import numpy as np
import pandas as pd
```

```python
from sklearn import datasets
```

```python
titanic = pd.read_csv("training_titanic_x_y_train.csv")
titanic_test = pd.read_csv("test_titanic_x_test.csv")
```

```python
titanic.describe()
```

```python
titanic.head()
```

```python
titanic.drop("Name", axis=1, inplace = True)
titanic.drop("Ticket", axis=1, inplace = True)
titanic.drop("Cabin", axis=1, inplace = True)
titanic.drop("Embarked", axis=1, inplace = True)

titanic_test.drop("Name", axis=1, inplace = True)
titanic_test.drop("Ticket", axis=1, inplace = True)
titanic_test.drop("Cabin", axis=1, inplace = True)
titanic_test.drop("Embarked", axis=1, inplace = True)
```

```python
titanic.head()
```

```python
titanic_test.head()
```

```python
titanic.Age.fillna(titanic.Age.mean(), inplace = True)
titanic_test.Age.fillna(titanic_test.Age.mean(), inplace = True)
titanic.describe()
```

```python
titanic_test.describe()
```

```python
for i in range(titanic.shape[0]):
    if titanic.Sex[i] == "male":
        titanic.Sex[i] = 0
    elif titanic.Sex[i] == "female":
        titanic.Sex[i]= 1

        

for i in range(titanic_test.shape[0]):
    if titanic_test.Sex[i] == "male":
        titanic_test.Sex[i] = 0
    elif titanic_test.Sex[i] == "female":
        titanic_test.Sex[i]= 1
 
```

```python
titanic.head()
## female = 1
## male = 0
```

```python
titanic_test.head()
```

```python
titanic.Sex = pd.to_numeric(titanic.Sex)
titanic_test.Sex = pd.to_numeric(titanic_test.Sex)
```

```python
titanic_test.dtypes
```

```python
titanic.head()
```

```python
from sklearn import preprocessing
```

```python
titanic_whole = pd.concat([titanic, titanic_test])
```

```python
del titanic_whole['Survived']
titanic_whole.describe()
```

```python
titanic_scaled = pd.DataFrame(preprocessing.scale(titanic_whole))
```

```python
titanic_scaled.describe()
```

```python
titanic_train_x = titanic_scaled.iloc[0:668,:]
titanic_test_x = titanic_scaled.iloc[668:891,:]

titanic_train_y = titanic.iloc[:,6]

titanic_train_x = titanic_train_x.values
titanic_test_x = titanic_test_x.values
titanic_train_y = titanic_train_y.values
```

```python
titanic_train_x
```

```python
titanic_test_x.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()
```

```python
clf.fit(titanic_train_x, titanic_train_y, c = 0.7)
```

```python
clf.score(titanic_train_x, titanic_train_y)
```

```python
output = clf.predict(titanic_test_x)
```

```python
np.savetxt("output.csv",output, fmt ='%1.0f')
```
