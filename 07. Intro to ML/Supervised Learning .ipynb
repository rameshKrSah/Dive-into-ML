```python
from sklearn import datasets
```

```python
boston = datasets.load_boston()

x = boston.data
y = boston.target
```

```python
from sklearn import model_selection 
x_train, x_test, y_train, y_test = model_selection.train_test_split(x,y)

print(x_train.shape)
print(x_test.shape)           #by default 3:1 . 25% for testing is good enough, but we can change the percentage
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression 
alg1 = LinearRegression()              #picked up the algorithm object/ specifier
```

```python
alg1.fit(x_train, y_train )            #training/learning from the algorithm 
```

```python
y_pred = alg1.predict(x_test)           #predicting the output after learning from the training data 
```

```python
# compare y_pred and y_test

import matplotlib.pyplot as plt
plt.scatter(y_test, y_pred)
plt.axis([0,40,0,40])
plt.show()
```
