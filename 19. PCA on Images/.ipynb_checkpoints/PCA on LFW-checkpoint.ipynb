```python
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn import ensemble

import time
```

```python
lfw = datasets.fetch_lfw_people(min_faces_per_person=100, resize=0.4)
```

```python
lfw.keys()
```

```python
lfw.data.shape
```

```python
lfw.images.shape
```

```python
# plotting first figsize=size=images
fig = plt.figure(figsize=(8,8))

for i in range(64):
    ax = fig.add_subplot(8, 8, i+1)
    ax.imshow(lfw.images[i], cmap = plt.cm.bone)
plt.show()
```

```python
x = lfw.data
y = lfw.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 0)
```

```python
pca = PCA()
pca.fit(x_train)
```

```python
k = 0
total = sum(pca.explained_variance_)
current_sum = 0

while(current_sum / total < 0.99):
    current_sum += pca.explained_variance_[k]
    k += 1
k
```

```python
# Whiten is true applies kind of feature scaling on image data 
pca = PCA(n_components=k, whiten=True)

transformed_data = pca.fit_transform(x_train)
transformed_data.shape
```

```python
x_approx = pca.inverse_transform(transformed_data)
x_approx.shape
```

```python
x_approx_images = x_approx.reshape((855, 50, 37))
```

```python
# plotting first 64 images
fig = plt.figure(figsize=(8,8))

for i in range(64):
    ax = fig.add_subplot(8, 8, i+1)
    ax.imshow(x_approx_images[i], cmap = plt.cm.bone)
plt.show()
```

# Eigen Faces

```python
eigen_vectors = pca.components_
eigen_vectors.shape
```

```python
eigen_faces = eigen_vectors.reshape((k,50,37))
```

```python
# plotting first 64 images
fig = plt.figure(figsize=(8,8))

for i in range(64):
    ax = fig.add_subplot(8, 8, i+1)
    ax.imshow(eigen_faces[i], cmap = plt.cm.bone)
plt.show()
```

```python
x_train_pca = transformed_data
x_test_pca = pca.transform(x_test)
```

# Random Forest without applying PCA

```python
rf = ensemble.RandomForestClassifier()
start_time = time.time()
rf.fit(x_train, y_train)
end_time = time.time()

print(end_time - start_time)

y_pred = rf.predict(x_test)

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

# Random Forest after applying PCA

```python
rf = ensemble.RandomForestClassifier()
start_time = time.time()
rf.fit(x_train_pca, y_train)
end_time = time.time()

print(end_time - start_time)

y_pred = rf.predict(x_test_pca)

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```
