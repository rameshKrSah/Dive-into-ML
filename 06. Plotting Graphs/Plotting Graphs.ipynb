```python
#customising graphs
# color, labels, marker, linewidth, labels for axes and datasets

import numpy as np

#x = np.array([1,2,3,4])
x = np.arange(0,5, 0.1)        #arange points from 0 to 4.9 with increment 0.1
y = x ** 3
y2 = x ** 2

#plt.plot(x,y,color= "black", marker = "+")

plt.plot(x,y,color= "black", linewidth = 6, label = "x^3")         
plt.plot(x,y2,color= "red", linewidth = 6, label = "x^2")         


plt.axis([0,10,0,200])        #axes specifiers [x_axis_si, x_axis_ei, y_axis_si, y_axis_ei] 
plt.grid()
plt.text(4,16, "Text", fontsize = 20)

plt.ylabel("y")
plt.xlabel("x")
plt.title("Graph of X cube: ")
plt.legend()

plt.show()
```

```python
#Pie Graphs

labels = ["A", "B", "C", "D"]
sizes = [2,4,3,6]
col = ["blue", "purple", "red", "yellow"]

expl = [0.1, 0,0.2,0]
plt.pie(sizes,explode = expl, colors = col, labels = labels, autopct = "%0.3f%%", counterclock = False, startangle =120)

#we can pass a function too for autopct i.e. labeling with some data
#By default counterclock is true
#By default startangle = 0

plt.title("Split among the classes")

plt.axis("equal")      # for two axes to be equal i.e. to make it a circle rather than an eclipse
plt.show()
```
